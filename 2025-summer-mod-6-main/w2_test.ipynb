```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Read the CSV file
df = pd.read_csv('homework_2.1.csv', index_col=0)

print("Data shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
```

```python

# Create long format data for regression
# Each row becomes 3 observations (one for each group)
long_data = []

for idx, row in df.iterrows():
    # Group 1 observation
    long_data.append({
        'time': row['time'],
        'outcome': row['G1'],
        'group1': 1,
        'group2': 0,
        'group3': 0
    })
    # Group 2 observation  
    long_data.append({
        'time': row['time'],
        'outcome': row['G2'],
        'group1': 0,
        'group2': 1,
        'group3': 0
    })
    # Group 3 observation
    long_data.append({
        'time': row['time'],
        'outcome': row['G3'],
        'group1': 0,
        'group2': 0,
        'group3': 1
    })

# Convert to DataFrame
long_df = pd.DataFrame(long_data)
long_df
```

```python

print(f"\nLong format data shape: {long_df.shape}")
print("\nFirst few rows of long format:")
print(long_df.head(9))

# Set up regression: outcome = β₀ + β₁*time + α₁*group1 + α₂*group2 + α₃*group3
X       = long_df[['time', 'group1', 'group2', 'group3']]
y       = long_df['outcome']

# Fit the regression model
model   = LinearRegression(fit_intercept=True)
model.fit(X, y)

# Extract coefficients
intercept   = model.intercept_
time_coef   = model.coef_[0]
group1_coef = model.coef_[1]
group2_coef = model.coef_[2]
group3_coef = model.coef_[3]
```

```python

print("\n" + "="*50)
print("REGRESSION RESULTS")
print("="*50)
print(f"Model: outcome = β₀ + β₁*time + α₁*group1 + α₂*group2 + α₃*group3")
print(f"\nIntercept (β₀):     {intercept:.6f}")
print(f"Time coefficient (β₁): {time_coef:.6f}")
print(f"Group 1 fixed effect (α₁): {group1_coef:.6f}")
print(f"Group 2 fixed effect (α₂): {group2_coef:.6f}")
print(f"Group 3 fixed effect (α₃): {group3_coef:.6f}")

print("\n" + "="*50)
print("ANSWER TO MULTIPLE CHOICE QUESTION")
print("="*50)
print(f"Group 1 coefficient: {group1_coef:.6f}")

# Compare to given options
options     = [0.01023, 0.00850, 0.00485, 0.1823]
differences = [abs(group1_coef - option) for option in options]

print(f"\nComparison to options:")
for i, (option, diff) in enumerate(zip(options, differences)):
    print(f"{i+1}. {option:.5f} - Difference: {diff:.6f}")

closest_idx = differences.index(min(differences))
print(f"\nClosest match: Option {closest_idx + 1} ({options[closest_idx]:.5f})")

# Calculate R-squared for model fit
y_pred      = model.predict(X)
ss_res      = np.sum((y - y_pred) ** 2)
ss_tot      = np.sum((y - np.mean(y)) ** 2)
r_squared   = 1 - (ss_res / ss_tot)
print(f"\nModel R-squared: {r_squared:.4f}")
```
