### 4. If ﻿W﻿ causes both ﻿X﻿ and ﻿Y﻿, and ﻿X﻿ causes ﻿Y﻿, then the influence of ﻿W﻿ tends to: 

Make the impact of X on Y appear stronger than it really is. 

Make the impact of X on Y appear weaker than it really is. 

<font color='gold'>It depends on the sign of the causal relationships involved.</font>

W is a collider, so it would not affect the impact of X on Y and should not be controlled for. 

*If W positively influences both X and Y, it will tend to increase the correlation. If it positively influences one and negatively influences the other, it will tend to decrease the correlation.*
