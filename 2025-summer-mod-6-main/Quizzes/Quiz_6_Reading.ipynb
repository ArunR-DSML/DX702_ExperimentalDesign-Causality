**a. `2`**

b. -4

c. 3

d. -2

Feedback
*The first treated item increases from 0 (counterfactual) to 3 (factual). All of the others have a smaller increase (or a decrease). For example, the first untreated item decreases from 1 (factual) to –3 (counterfactual). The treated items move from counterfactual (untreated) to factual (treated), while the untreated items move from factual (untreated) to counterfactual (treated), given treatment.*
